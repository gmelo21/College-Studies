```python
from scipy.stats import norm, binom

mu = 5
sigma = 0.2

prob_normal = 1 - norm.cdf(5.3, mu, sigma)
print(f"{prob_normal:.3f}")
```

```python
mu = 120
sigma = 15

prob_normal = norm.cdf(100, mu, sigma)
print(f"Chance of finishing the exam in 100 minutes: {prob_normal:.3f}")

prob_normal = 1 - norm.cdf(150, mu, sigma)
print(f"Chance of finishing the exam in more than 150 minutes: {prob_normal:.3f}")

prob_normal1 = norm.cdf(110, mu, sigma)
prob_normal2 = norm.cdf(140, mu, sigma)
print(f"Chance of finishing the exam between 110 and 140 minutes: {(prob_normal2 - prob_normal1):.3f}")

percentage = norm.ppf(0.95, mu, sigma)
print(f"Minimum duration so that 95% of all students finish: {percentage:.3f}")

percentage = norm.ppf(0.2, mu, sigma)
print(f"Maximum duration so that 20% of the fastest students finish: {percentage:.3f}")
```
