```python
import pandas as pd
df = pd.read_csv('/content/data_sw.csv')
```

```python
# List the first 5 characters from the DataFrame.

df.head(5)
```

```python
# What are the unique species in the dataset?

df['Espécie'].value_counts()
```

```python
# What is the average Force Power by Species?

df.groupby('Espécie')['Poder_Forca'].mean().sort_values(ascending=False)
```

```python
# Use groupby() to calculate the average Force Power by Planet and determine where the most powerful characters come from.

df.groupby('Planeta')['Poder_Forca'].mean().sort_values(ascending=False)
```

```python
# Filter the characters who have a Force Power greater than 50.

df[df['Poder_Forca'] > 50]
```

```python
# Create a new column called "Power Category" based on the Force Power level.

df['Categoria_De_Poder'] = df['Poder_Forca'].apply(lambda x: 'Jedi Master' if x > 80 else 'Sem Sensibilidade' if x < 30 else 'Padawan')
df.head()
```

```python
# Sort the DataFrame by Combat Skill level in descending order.

df.sort_values('Habilidade_Combate', ascending=False)
```

```python
# Find the character with the highest Combat Skill.

df[df['Habilidade_Combate'] == df['Habilidade_Combate'].max()]
```

```python
# Apply a style to the DataFrame to highlight in red the characters with Force Power greater than 80.

def highlight_name(value):
    return 'font-weight: bold; color: red' if (df[df['Personagem'] == value]['Poder_Forca'] > 80).any() else ''

df.style.map(highlight_name, subset=['Personagem'])
```

```python
# What is the average age of characters at the time of the Battle of Yavin (Year 0)?

df_filtered_2 = df[df['Ano_Nascimento'] <= 0]
idade = -(df_filtered_2['Ano_Nascimento'].mean())
print(f"{idade:.2f}")
```
