```python
from scipy.stats import expon, norm, binom
import pandas as pd

lmbda = 0.5
prob_expo = expon.cdf(2, scale=1/lmbda)
print(f"Probability of churning in, at most, 2 seconds: {prob_expo:.3f}")
```

```python
df = pd.read_csv("dataset_probability.csv")

df.head()
```

```python
# 1A
mean = df["tempo_espera"].mean()
print(f"{mean:.3f}")
```

```python
# 1B
lmbda = 1 / mean
print(f"{lmbda:.3f}")
```

```python
# 1C
prob_expo = expon.cdf(3, scale=1/lmbda)
print(f"{prob_expo:.3f}")
```

```python
# 2A
mean = df["avaliacoes"].mean()
print(f"Mean: {mean:.3f}")
std = df["avaliacoes"].std()
print(f"STD: {std:.3f}")
```

```python
# 2B
lower = norm.ppf(0.025, mean, std)
upper = norm.ppf(0.975, mean, std)
print(f"Lower limit: {lower:.3f}\nUpper limit: {upper:.3f}")
```

```python
# 2C
print("É o intervalo dos 2.5% menores e maiores avaliações.")
```

```python
# 3A
n = 10
p = 0.6

prob_binom = binom.pmf(8, n, p)
print(f"{prob_binom:.3f}")
```

```python
# 3B
sumAll = 0
for i in range(7, 11):
    prob_binom = binom.pmf(i, n, p)
    sumAll += prob_binom
print(f"{sumAll:.3f}")
```

```python
# Extra 1
p = 0.6
n = 24

missable = int((25/100 * 24) + 1)
sumAll = 0

for i in range(missable, 24):
    prob_binom = binom.pmf(i, n, p)
    sumAll += prob_binom

print(f"{sumAll:.3f}")
```
