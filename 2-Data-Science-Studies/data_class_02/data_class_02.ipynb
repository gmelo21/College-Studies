```python
# 1-10 - Setting the heroes dataframe as our main one:
df = pd.read_csv('/content/data_heroes.csv')

# Printing the whole heroes dataframe:
df
```

```python
# 2 - Who is the oldest hero on the list?

df.loc[df['Ano_Estreia'] == df['Ano_Estreia'].min()][['Nome', 'Ano_Estreia']]
```

```python
# 3 - What is the average strength of the DC heroes?

df[df['Editora'] == 'DC']['For√ßa'].mean()
```

```python
# 8 - List the heroes in descending order of speed.

df.sort_values('Velocidade', ascending=False)[['Nome', 'Velocidade']]
```

```python
# 11-20 - Setting the team dataframe as our main one:
df = pd.read_csv('/content/data_team.csv')

# Printing the whole team dataframe:
df
```

```python
# 13 - What is the average final position of Corinthians in the championships?

df['Posi√ß√£o_Campeonato'].mean()
```

```python
# 14 - Which year had the fewest losses?

df[df['Derrotas'] == df['Derrotas'].min()][['Ano', 'Derrotas']]
```

```python
# 15 - In which years did Corinthians score more than 60 goals?

df[df['Gols_Marcados'] > 60][['Ano', 'Gols_Marcados']]
```

```python
# 17 - List the years in descending order of points (Wins = 3 points, Draws = 1 point).

df[['Ano']].assign(Pontos = (df['Vit√≥rias'] * 3) + (df['Empates'] * 1)).sort_values(by='Pontos', ascending=False)
```

```python
# 18 - What was the worst position of Corinthians in this period?

df[df['Posi√ß√£o_Campeonato'] == df['Posi√ß√£o_Campeonato'].max()][['Ano', 'Posi√ß√£o_Campeonato']]
```

```python
# 19 - What year had the most games played?

df[df['Jogos'] == df['Jogos'].max()][['Ano', 'Jogos']]
```

```python
# 20 - What is the average goals taken per game in each year?

df[['Ano']].assign(M√©dia_Gols_Sofridos = df['Gols_Sofridos'] / df['Jogos']).round(2)
```
