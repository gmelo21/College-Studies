```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y, color='red')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='orange')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter plot')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [4, 7, 1, 8, 5]

plt.bar(categories, values, color='green')
plt.show()
```

```python
values = np.random.randn(1000)

plt.hist(values, bins = 20, edgecolor = 'black', color='cyan')
plt.show()
```

```python
values = [4, 5, 12, 16, 25]

plt.boxplot(values)
plt.show()
```

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
hours = ['Morning', 'Lunch', 'Afternoon', 'Night', 'Late Night', 'Dark Hour']
sales = np.random.randint(0, 50, (7, 6))

plt.figure(figsize = (8, 5))
sns.heatmap(sales, annot = True, cmap = 'coolwarm', xticklabels = hours, yticklabels = days)

plt.xlabel('Period')
plt.ylabel('Weekday')
plt.title('Heatmap')
plt.show()
```

```python
plt.plot(x, y, color='blue', linestyle='--', marker='s', label='Data Series')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.legend()
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

ax[0].plot(x, y, color='blue')
ax[1].bar(categories, values, color='green')
plt.show()
```
